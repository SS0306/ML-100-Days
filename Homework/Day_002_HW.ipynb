```python
train_2nd=app_train.head(100)
age_for_client=round(abs((app_train.DAYS_BIRTH.head(100))/365.25))
train_2nd['age_for_client']=age_for_client
```

```python
#scatter plot:以前100筆為例，用scatter plot畫出年齡和貸款之間的相關性
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(train_2nd.age_for_client,train_2nd.AMT_CREDIT)
plt.xlabel('Age')
plt.ylabel('AMT_CREDIT')
plt.title('Correlation of Age and Credit Amount--n=100')
plt.show
```
